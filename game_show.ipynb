```python
perguntas = {
    'Pergunta 1': {
        'Pergunta': 'Quanto 2 + 2? ', 
        'Respostas': {'a': '2', 'b': '3', 'c': '4', 'd': '5'},
                    'resposta_certa': 'c',
    },
    
    'Pergunta 2': {
        'Pergunta': 'Qual a capital de Madagascar? ', 
        'Respostas': {'a': 'Quito', 'b': 'Lima', 'c': 'Camoboja', 'd': 'Antananarivo'},
                    'resposta_certa': 'd',
    },
    
    'Pergunta 3': {
        'Pergunta': 'Em que país está localizado o vulcão mais antigo do mundo? ', 
        'Respostas': {'a': 'EUA', 'b': 'Brasil', 'c': 'Indonésia', 'd': 'China'},
                    'resposta_certa': 'b',
    },
    
    'Pergunta 4': {
        'Pergunta': 'Quantos anos durou a Guerra dos 100 anos? ', 
        'Respostas': {'a': '100', 'b': '116', 'c': '99', 'd': '101'},
                    'resposta_certa': 'b',
    },
    
    'Pergunta 5': {
        'Pergunta': 'Quem é considerado o pai da computação? ', 
        'Respostas': {'a': 'Steve Jobs', 'b': 'Bill Gates', 'c': 'Alan Turing', 'd': 'Elon Musk'},
                    'resposta_certa': 'c',
    },
    
    'Pergunta 6': {
        'Pergunta': 'Quanto tempo a luz do Sol leva para chegar na Terra? ', 
        'Respostas': {'a': '10 minutos', 'b': '5 minutos', 'c': '2 minutos', 'd': '8 minutos'},
                    'resposta_certa': 'd',
    },
    
    'Pergunta 7': {
        'Pergunta': 'Quem inventou a linguagem de programação Python? ', 
        'Respostas': {'a': 'Guido van Rossum', 'b': 'Guy L. Steele Jr.', 'c': 'James Gosling', 'd': 'Bill Joy'},
                    'resposta_certa': 'a',
    },
    
}

respostas_certas = 0

print('Atenção. Qualquer alternativa que não estiver nas opções será desconsiderada.')
print('-' * 78)

for pk, pv in perguntas.items():
    print(f'{pk}: {pv["Pergunta"]}')
    
    print('Alternativas:')
    
    for rk, rv in pv['Respostas'].items():
        print(f'{rk}. {rv}')
        
    resposta_usuario = str(input('Sua resposta: ')).lower().strip()[0]
    
    if resposta_usuario not in pv['Respostas']:
        print('Resposta Inválida!')
        
    elif resposta_usuario == pv['resposta_certa']:
        print('Resposta Correta!')
        respostas_certas += 1
    else:
        print('Que pena, você errou!')
        
    print()
    
porcentagem_acerto = (respostas_certas / len(perguntas)) * 100
print(f'Porcentagem de acerto: {porcentagem_acerto:.0f}%')

if porcentagem_acerto >= 80:
    print('Parabéns! Você é muito esperto.')
elif 50 <= porcentagem_acerto < 80:
    print('Muito bem. Mas dá pra melhorar.')
elif 30 <= porcentagem_acerto < 50:
    print('Razoável!!')
else:
    print('Que pena, seu resultado foi péssimo.')
```
